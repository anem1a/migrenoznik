### Коды ошибок в проекте "Мигренозник"

__0__: ошибки нет                   
__1__: пользователь с таким логином уже есть                      
__2__: пароль слишком простой                 
__3__: логин содержит недопустимые символы                            
__4__: логин или пароль пуст                              
__5__: другая ошибка                                 
__13__: пользователь не авторизован                               
__444__: ошибка вводных данных (валидация)                                      
__666__: ошибка сервера, бд
